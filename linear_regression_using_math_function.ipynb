```python
def loss_function(m,b,points):
    total_error = 0
    for i in range(len(points)):
        x = points.iloc[i].studytime
        y = points.iloc[i].score
        total_error += (y - (m*x + b)) ** 2
    total_error / float(len(points))
```

```python
def gradient_descent(m_now,b_now,points,L):  # L is learning rate
    m_gradient = 0
    n_gradient = 0

    n = len(points)

    for i in range(n):
        x = points.iloc[i].studytime
        y = points.iloc[i].score

        m_gradient += -(2/n) * x * (y - (m_now * x + b_now))
        b_gradient += -(2/n) * (y - (m_now * x + b_now))
# for actually minimizing the equation using derivatives above and learning rate below
    m = m_now - m_gradient * L
    b = m_now - b_gradient * L
    return m,b
```

```python

# Gradient descent function
def gradient_descent(m, b, data, L):
    N = len(data)
    m_gradient = 0
    b_gradient = 0
    for i in range(N):
        x = data['studytime'][i]
        y = data['score'][i]
        m_gradient += -2/N * x * (y - (m * x + b))
        b_gradient += -2/N * (y - (m * x + b))
    m -= L * m_gradient
    b -= L * b_gradient
    return m, b
```

```python


# Initialize parameters
m = 0
b = 0
L = 0.0001
epochs = 1000

# Perform gradient descent
for i in range(epochs):
    m, b = gradient_descent(m, b, data, L)

print(m, b)
```

```python
# Plot the data and the regression line
plt.scatter(data['studytime'], data['score'], color="blue")
plt.plot(list(range(0, 12)), [m * x + b for x in range(0, 12)], color="red")
plt.xlabel('Study Time')
plt.ylabel('Exam Score')
plt.show()
```
